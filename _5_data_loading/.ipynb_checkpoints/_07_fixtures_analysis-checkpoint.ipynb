### Load data

```python
# load the data from psql
df = pd.read_sql_query("SELECT * FROM " + schema1 + ".match_fixtures",db)
```

```python
# Drop columns without data
df = df.drop({'index','stage','attendance'},axis=1)
```

```python
# Plot scores of the different teams by the match number
df = df.sort_values(by=['tournament_id','matchid'],ascending = True)
```

```python
# We set a number of columns to be integers
int_col = ['matchnumber','team1score','team2score']
for k in int_col:
    df[k] = df[k].astype(int)
```

```python
# Select all games played by South Africa
saf1 = df[df['team1id']=='2420']
saf1 = saf1[['eventname','matchnumber','team1score']]
#saf1 = saf1.groupby(['eventname'])['team1score'].agg(['count','sum'])
saf2 = df[df['team2id']=='2420']
saf2 = saf2[['eventname','matchnumber','team2score']]
#saf2 = saf2.groupby(['eventname'])['team2score'].agg(['count','sum'])
saf = pd.merge(left = saf1, right = saf2, on='eventname', how='left')
saf= saf.rename(columns = {'matchnumber_x':'matchnumber_team1','matchnumber_y':'matchnumber_team2'})#.reset_index(drop=False)
saf.head()
```
