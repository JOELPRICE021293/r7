```python
# Set up connection with the database
def getDatabase(db):
    # http://initd.org/psycopg/docs/usage.html
    connection = psycopg2.connect(database=db, user="postgres", password="Oxford16")
    connection.set_isolation_level(ISOLATION_LEVEL_AUTOCOMMIT)
    return connection
```

```python
# Run the query and submit to the database
def submitToDatabase(query, query_values, connection):
    answer = connection.cursor()
    answer = answer.execute(query,(query_values))
    return answer
```

```python
# Normalize and clean text contained in tweets
def normalizeText(tweet_text):
    str_text = tweet_text
    normalizedText = '\'' + str_text.replace("'", r"''") + '\''
    return normalizedText
```
