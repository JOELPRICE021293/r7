### Reading Match Fixtures into datafrmae

step 1 is to extract the xml from the web to an xml file saved locally using the command line:

wget http://webservices.irb.com/EventInformation.asmx/FixturesResults?UID=e9656db8-ffb5-4115-ac0b-cbd5688e6648&EventID=1684" --output-document=try_Fixtures_results_1684.xml

```python
# Step 2: we run the function
def xml2df(xml_data):
    root = ET.XML(xml_data) # element tree
    all_records = []
    for i, child in enumerate(root):
        record = {}
        for subchild in child:
            record[subchild.tag] = subchild.text
            all_records.append(record)
    return pd.DataFrame(all_records)
```

```python
# Define the input and output strings
input_string = '../_4_data_extracts/fixtures_151617/'
output_strig = '../_6_data_clean/'
filename = 'try_Fixtures_results_1611.xml'
```

```python
# Read the filenames in the folder into a dicitionary
arr_files = glob.glob(input_string + "**/*.xml", recursive=True)
#arr_files
```

```python
# Define an empty dataframe to append the different xml files to
full_fixtures = pd.DataFrame()

# We create a loop where we open the different xml extracts in a pandas dataframe, clean the data and append it to the
# full fixtures dataframe. 
for k in arr_files:
    # Set the data object to be the xml file
    xml_data = open(k).read()
    
    # Create a df which equals the data object
    df = xml2df(xml_data)
    
    # Clean the column headers 
    dict_columns={}
    for x in (df.columns.values):
        dict_columns[x] = x.replace('{http://webservices.irb.com/}','')
    df_clean = df.rename(columns=dict_columns)
    
    # remove duplicates from the data
    df_clean = df_clean.drop_duplicates(['MatchID'], keep='first')
    
    # Append each individual dataframe to the full_fixtures df
    full_fixtures = pd.concat([full_fixtures, df_clean],axis=1)
```
