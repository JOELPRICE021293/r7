##### Approach that works for simpler xml format

Problem: for the format I'm working with it does not read in the information of the subchildren after the first one due to the different format

```python
# The below approach works for the squad file, but not for the match file. I.e. for the match file I do get the 
# information linked to the 'MatchInformation', but not for 'Officials', 'Team1Sheet', 'Team2Sheet' and 'Timeline'. 

xml_data = open('../_4_data_extracts/Test/squad_1685_2408.xml').read()

#xml_data = open('../_4_data_extracts/Test/match_1693_24947.xml').read()

# Step 2: we run the function
def xml2df(xml_data):
    root = ET.XML(xml_data) # element tree
    all_records = []
    for i, child in enumerate(root):
        record = {}
        for subchild in child:
            record[subchild.tag] = subchild.text
            all_records.append(record)
    return pd.DataFrame(all_records)

df = xml2df(xml_data)
# The output df has some duplicates which we drop.
df = df.drop_duplicates(['{http://webservices.irb.com/}PersonID'], keep='first')
df
```
