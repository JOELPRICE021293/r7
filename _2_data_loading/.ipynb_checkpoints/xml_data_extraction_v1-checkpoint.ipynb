### Reading Match Fixtures into PostgreSQL

step 1 is to extract the xml from the web to an xml file saved locally using the command line:

wget http://webservices.irb.com/EventInformation.asmx/FixturesResults?UID=e9656db8-ffb5-4115-ac0b-cbd5688e6648&EventID=1684" --output-document=try_Fixtures_results_1684.xml

```python
# we open the xml file we just created
xml_data = open('../_4_data_extracts/Test/Match_1684.xml').read()

def xml2df(xml_data):
    root = ET.XML(xml_data) # element tree
    all_records = []
    for i, child in enumerate(root):
        record = {}
        for subchild in child:
            record[subchild.tag] = subchild.text
            all_records.append(record)
    return pd.DataFrame(all_records)
```

```python
# We read teh xml file into a pandas dataframe
df = xml2df(xml_data)
dict_columns={}
for x in (df.columns.values):
            dict_columns[x] = x.replace('{http://webservices.irb.com/}','')
df_clean = df.rename(columns=dict_columns)
df_clean = df_clean.drop_duplicates(['MatchID'], keep='first')
df_clean.head()
```
